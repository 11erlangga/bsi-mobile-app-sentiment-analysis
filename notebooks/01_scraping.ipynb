Pada proyek ini dilakukan **scraping data review aplikasi Byond by BSI** yang kemudian akan dianalisis menggunakan **sentiment analysis** untuk mengidentifikasi kecenderungan sentimen pengguna.

# Import Dependencies

```python
# Install library untuk scraping review dari Google Play
!pip install google-play-scraper -q
```

```python
from google_play_scraper import Sort, reviews
import pandas as pd
```

# Scraping The Reviews

Parameter yang digunakan pada fungsi `reviews()` adalah sebagai berikut:
* lang='id' → hanya mengambil review berbahasa Indonesia
* country='id' → review dari region Indonesia
* sort=Sort.NEWEST → urutkan berdasarkan review terbaru
*  count=20000 → jumlah review yang ingin diambil

```python
# Scraping seluruh ulasan aplikasi BSI Mobile (Bank Syariah Indonesia)
# ID aplikasi: com.bsm.activity2
review_list, continuation_token = reviews(
    'com.bsm.activity2',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=20_000
)
```

# Convert to DataFrame

Hasil scraping yang diperoleh dari fungsi `reviews()` masih berbentuk **list of dictionary**. Agar data lebih mudah diproses dan dianalisis menggunakan Python, data tersebut diubah ke dalam bentuk **pandas DataFrame**.

```python
# Mengubah hasil scraping (list of dictionary) menjadi DataFrame
app_reviews_df = pd.DataFrame(review_list)
app_reviews_df.head()
```

Method `info()` digunakan untuk melihat ringkasan dataset seperti jumlah baris data, jumlah kolom, serta tipe data dari masing-masing kolom.

```python
# Melihat informasi dataset
app_reviews_df.info()
```

Dari proses scraping tersebut berhasil diperoleh **20,000 data ulasan pengguna** sesuai dengan jumlah yang diminta pada parameter `count`.

Untuk memastikan kualitas dataset, dilakukan penghapusan data duplikat menggunakan method `drop_duplicates()`.

```python
# Menghapus data review yang duplikat
app_reviews_df.drop_duplicates(inplace=True)
```

# Save Reviews

Dataset kemudian disimpan ke dalam file **CSV** agar dapat digunakan kembali pada tahap **labelling, pre-processing dan analisis sentimen** tanpa perlu melakukan proses scraping ulang.

```python
# Menyimpan kolom-kolom penting ke file CSV
app_reviews_df[
    [
        "reviewId",
        "content",
        "score",
        "thumbsUpCount",
        "reviewCreatedVersion",
        "at",
        "appVersion",
    ]
].to_csv("review_bsi_mobile.csv", index=False)
```
